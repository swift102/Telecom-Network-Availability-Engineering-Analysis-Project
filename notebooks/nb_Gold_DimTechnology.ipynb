```python
# --------------------------------------------
# Insert and write to gold table
# --------------------------------------------
from pyspark.sql.functions import *

silver_df = spark.table("lh_Silver_Telecom.dbo.silver_network_events")

dim_technology = (
    silver_df
    .select(col("Technology").alias("technology_name"))
    .dropna()
    .dropDuplicates()
    .withColumn("technology_key", monotonically_increasing_id())
)

(
    dim_technology
    .write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("lh_Gold_Telecom.dbo.dim_technology")
)

print("✓ dim_technology created")
```
