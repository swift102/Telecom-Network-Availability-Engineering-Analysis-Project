```python
# --------------------------------------------
# Insert and write to gold table
# --------------------------------------------
from pyspark.sql.functions import *

silver_df = spark.table("lh_Silver_Telecom.dbo.silver_network_events")


dim_province = (
    silver_df
    .select(
        col("Province_Code").alias("province_code"),
        col("Province").alias("province_name")
    )
    .dropna()
    .dropDuplicates()
    .withColumn("province_key", monotonically_increasing_id())
)

(
    dim_province
    .write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("lh_Gold_Telecom.dbo.dim_province")
)

print("✓ dim_province created")
```
